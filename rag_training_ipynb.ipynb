```python
knowledge = ""

files = [
    "clinic.md",
    "services.md",
    "faq.md",
    "appointment.md",
    "emergency.md",
    "safety.md"
]

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        knowledge += f.read() + "\n\n"

print(knowledge)
```

```python
print(len(knowledge))
```

```python
chunk_size = 500

chunks = []

for i in range(0, len(knowledge), chunk_size):
    chunks.append(knowledge[i:i+chunk_size])

print("Number of Chunks:", len(chunks))
```

```python
print(chunks[0])
```

```python
print(chunks[1])
```

```python
print("Total Chunks:", len(chunks))

for i, chunk in enumerate(chunks):
    print(f"\n----- Chunk {i+1} -----\n")
    print(chunk)
```
